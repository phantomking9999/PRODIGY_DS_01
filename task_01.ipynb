**4. Visualization :**

we first prepare data :

```python
# we prepare data

tot_d = tot.drop(columns=['Country Name', 'Indicator Name', 'Indicator Code'])
fe_d = fe.drop(columns=['Country Name', 'Indicator Name', 'Indicator Code'])
ma_d = ma.drop(columns=['Country Name', 'Indicator Name', 'Indicator Code'])
```

```python
# re-structure data
tot_n = pd.melt(tot_d, id_vars=["Country Code"], var_name="year", value_name="total_population")
fe_n = pd.melt(fe_d, id_vars=["Country Code"], var_name="year", value_name="female_population")
ma_n = pd.melt(ma_d, id_vars=["Country Code"], var_name="year", value_name="male_population")
```

```python
# ex of verification
tot_n.head()
```

```python
# fuse data
data = tot_n.merge(fe_n, on=["Country Code", "year"])
data = data.merge(ma_n, on=["Country Code", "year"])
```

```python
# ex of verification
data.head()
```

```python
# Calculate proportions
data["fe_part"] = data["female_population"] / data["total_population"]
data["ma_part"] = data["male_population"] / data["total_population"]
```

The idea is to visualize evolution of gender distribution for a country over years

```python
country_code = "YEM"

country_data = data[data["Country Code"] == country_code]

# Convert years to int so the can be ordered
country_data["year"] = country_data["year"].astype(int)
country_data = country_data.sort_values("year")

# visualization
plt.figure(figsize=(12, 6))
plt.bar(country_data["year"], country_data["fe_part"], label="Female", color="purple")
plt.bar(country_data["year"], country_data["ma_part"], bottom=country_data["fe_part"], label="Male", color="blue")
plt.title(f"Evolution of Gender Distribution over years for {tot[tot["Country Code"] == country_code].iloc[0,0]}")
plt.xlabel("Year")
plt.ylabel("Gender Distro")
plt.legend()
plt.show()
```
